```python
from keras.applications.vgg16 import VGG16
model = VGG16()
```

```python
model.summary()
```

```python
from keras.utils  import plot_model
plot_model(model)
```

```python
for i in range(len(model.layers)):
  # Check for convoliution layer
  if 'conv' not in model.layers[i].name:
    continue
  # Get filter names
  filters, biases = model.layers[i].get_weights()
  print("layer number",i,model.layers[i].name,filters.shape)
```

```python
# Retrieve weights from second hidden layer
filters, biases = model.layers[1].get_weights()
print(filters.shape)
```

```python
# Normalizing filter values to visualize them

f_min, f_max = filters.min(), filters.max()
filters = (filters - f_min) / (f_max - f_min)
```

```python
import matplotlib.pyplot as pyplot
n_filters = 6
ix = 1
fig = pyplot.figure(figsize=(20,20))
for i in range(n_filters):
  # Get the filters
  f = filters[:,:,:,i]
  for j in range(3):
    # Subplot for 6 filters
    pyplot.subplot(n_filters,3,ix)
    pyplot.imshow(f[:,:,j],cmap = 'grey')
    ix += 1
# Show the figure
pyplot.show()
```

```python
from keras.models import Model
model = Model(inputs = model.inputs, outputs = model.layers[1].output)
```

```python
from tensorflow.keras.utils import load_img, img_to_array
from tensorflow.keras.applications.vgg16 import preprocess_input
import numpy as np
from numpy import expand_dims

image = load_img("/content/WhatsApp Image 2025-03-06 at 18.00.25_192bdaac.jpg",target_size = (224,224))
# Convert the image to an array
image = img_to_array(image)
# Expand dimensions so that it represents a single 'sample'

image = expand_dims(image,axis = 0)
# Prepare the image for the VGG model
image = preprocess_input(image)
```

```python
# Calculating feature maps

feature_maps = model.predict(image)

fig = pyplot.figure(figsize=(20,15))

for i in range(1,feature_maps.shape[3]+1):
  pyplot.subplot(8,8,i)
  pyplot.imshow(feature_maps[0,:,:,i-1],cmap = 'gray')
pyplot.show()
```

```python
model2 = VGG16()
```

```python
layer_index = [2, 5, 9, 13, 17]
outputs = [model2.layers[i].output for i in layer_index]
model3 = Model(inputs = model2.inputs, outputs = outputs)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Predict feature maps
feature_maps = model3.predict(image)

# Iterate over selected layers and their corresponding feature maps
for layer_idx, feature_map in zip(layer_index, feature_maps):
    num_filters = feature_map.shape[3]  # Number of filters in the feature map
    grid_size = int(np.ceil(np.sqrt(num_filters)))  # Ensures a square grid layout

    fig, axes = plt.subplots(grid_size, grid_size, figsize=(15, 12))
    fig.suptitle(f"Feature Maps - Layer {layer_idx}", fontsize=20)

    # Flatten axes for easy indexing
    axes = np.array(axes).flatten()

    for i in range(num_filters):
        axes[i].imshow(feature_map[0, :, :, i], cmap='gray')
        axes[i].axis('off')

    # Hide unused subplots
    for j in range(num_filters, len(axes)):
        axes[j].axis('off')

    plt.show()

```
